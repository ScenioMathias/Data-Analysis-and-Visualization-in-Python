Data Visualization in Python

This material demonstrates some ways to visualize data in Python using Matplotlib and Seaborn for graph visualization.

To get started, let's first import the libraries.

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

Generating Histogram Plot in Matplotlib

The histogram allows us to analyze how a specific variable is distributed. To plot this graph, it is necessary to provide a vector or dataframe containing the x and y columns. In our case, we want to visualize the distribution of the 'Height' variable, but our dataset has 31 rows, which would make the graph very detailed. For a better visualization, we need to summarize the information into intervals. To do this, we will follow the process below.

```python
# Generating a histogram with 6 bins for the second column of the dataset
h = np.histogram(base.iloc[:,1], bins=6)

# Displaying the histogram
h
```

Plotting histogram charts below.

```python
# Plotting a histogram with 6 bins for the second column of the dataset
plt.hist(base.iloc[:,1], bins=6)

# Adding title and axis labels to the plot
plt.title('Arvores')
plt.ylabel('Frequencia')
plt.xlabel("Altura")
```

Another way to create histograms in Python is through the Seaborn library, as shown below.

```python
# Plotting a histogram with 6 bins for the second column of the dataset using seaborn
sns.histplot(base.iloc[:,1], kde=False, bins=6, color='blue').set(title='Arvores')
```

Now we will plot a density graph using the Seaborn library. The density graph has a function similar to the histogram, but it presents a continuous and smoothed curve.

```python
# Plotting a kernel density estimate plot for the second column of the dataset using seaborn
sns.kdeplot(base.iloc[:,1], color='blue').set(title='arvores')
```

Now, see both graphs overlaid on each other.

```python
# Plotting a histogram with kernel density estimation for the second column of the dataset using seaborn
sns.histplot(base.iloc[:,1], kde=True, bins=6, color='blue').set(title='Arvores')
```

Now we will generate a scatter plot using the Matplotlib library. This plot allows us to analyze two continuous variables and observe their relationships. See below the process to generate it.

```python
# Creating a scatter plot with 'Girth' on the x-axis and 'Volume' on the y-axis
plt.scatter(x=base.Girth, y=base.Volume, color='blue', facecolors='none', marker='*')

# Adding title and axis labels to the plot
plt.title('Arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```


Another example of plotting, with default markers.

```python
# Creating a scatter plot with 'Girth' on the x-axis and 'Volume' on the y-axis
plt.scatter(x=base.Girth, y=base.Volume, color='blue', facecolors='none')

# Adding title and axis labels to the plot
plt.title('Arvores')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```


It's also possible to generate the same plot using the Seaborn library, in this case, we can notice the smoother graphs.

Scatter plot with legends, allowing for differentiation between two variables.

Importing a new dataset.

```python
# Loading dataset from 'co2.csv' file and checking its dimensions
base = pd.read_csv('co2.csv')

# Displaying the dataset
base.shape
base
```

Assigning numeric columns to variables.

```python
# Assigning 'conc' column to variable x and 'uptake' column to variable y
x = base.conc
y = base.uptake
```


Assigning unique values from the Treatment attribute.

```python
# Creating a list of unique treatments from the 'Treatment' column
unico = list(set(base.Treatment))

# Plotting a scatter plot with x and y, labeling each point with its corresponding treatment
plt.scatter(x, y, label=unico)

# Adding a legend to the plot
plt.legend(loc='lower right')

```

The for loop creates separate scatter plots for each unique treatment in the base DataFrame.

```python
# Plotting scatter plots for each unique treatment
for i in range(len(unico)):
    indice = base.Treatment == unico[i]
    plt.scatter(x[indice], y[indice], label=unico[i])

# Adding a legend to the plot
plt.legend(loc='lower right')
```

Bar chart and Pie chart. Bar charts are great for categorized data, where the X-axis represents the category and the Y-axis represents its value. Pie charts represent data in percentages, in a circular format, and should be used when dealing with percentage data where the sum of all parts equals 100%.

Pie chart.

```python
# Plotting a pie chart of the grouped data
agrupado.plot.pie()
# Plotting a pie chart of the grouped data with legend
agrupado.plot.pie(legend=True)
```


Scatter Plot with Seaborn

It's also possible to generate scatter plots using the Seaborn library.


Importing dataset.


```python
# Loading the dataset
base = pd.read_csv('co2.csv')

# Displaying the first few rows of the dataset
base.head()
```


However, we can separate the legends in a more practical way, as shown below.

```python
# Scatter plot using 'conc' and 'uptake' attributes, grouped by 'Type'
srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')
plt.show()
```

Now, let's consider another approach. If we separate the searches by type, we can generate two graphs in a single figure, each referencing a different type.

```python
# Selecting specific records from the dataset (Quebec and Mississippi)
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
# Creating a figure with two subplots
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)

# Scatter plot for Quebec
srn.scatterplot(x=q.conc, y=q.uptake)
plt.title('Quebec')

# Scatter plot for Mississippi
plt.subplot(1, 2, 2)
srn.scatterplot(x=m.conc, y=m.uptake)
plt.title('Mississippi')

# Adjusting layout
plt.tight_layout()

# Displaying the plot
plt.show()
```


The same now, but separating by 'Treatment'.

```python
# Selecting records for chilled and non-chilled treatments
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
# Creating a figure with two subplots
plt.figure(figsize=(10, 5))

# Subplot for chilled treatment
plt.subplot(1, 2, 1)
srn.scatterplot(x=ch.conc, y=ch.uptake)
plt.title('Chilled')

# Subplot for non-chilled treatment
plt.subplot(1, 2, 2)
srn.scatterplot(x=nc.conc, y=nc.uptake)
plt.title('Non chilled')

# Adjusting layout
plt.tight_layout()

# Displaying the plot
plt.show()
```

With the Seaborn library, we can also generate scatter plots with categorical data using the 'catplot' command instead of 'scatterplot'. Below is an example where we selected two categorical columns to analyze.

```python
# Loading another dataset, esophagus cancer
base = pd.read_csv('esoph.csv')

# Displaying the dataset
base
```

```python
# Creating a categorical plot between 'alcgp' and 'ncontrols'
srn.catplot(x='alcgp', y='ncontrols', data=base, jitter=False)
plt.show()
```

Here's another approach where we generate multiple graphs according to each category existing in the 'tobgp' column.

```python
# Creating a categorical plot between 'alcgp' and 'ncontrols' with grouping by 'tobgp'
srn.catplot(x='alcgp', y='ncontrols', data=base, col='tobgp')
plt.show()
```

3D Scatter Plot. 

In a 3D Scatter Plot, an additional variable emerges with the advent of the new dimension.


Importing dataset.

```python
# Loading dataset from 'orchard.csv' file
base = pd.read_csv('orchard.csv')

# Displaying the dataset
base
```

```python
# Creating a 3D scatter plot
figura = plt.figure()
eixo = figura.add_subplot(1, 1, 1, projection='3d')
eixo.scatter(base.decrease, base.rowpos, base.colpos)

# Setting labels for each axis
eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')
```
